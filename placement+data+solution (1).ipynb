### Question 1

#### Check if your data set contains any null or empty values by defining a function named missing_values. The function should return a dataframe with the no. of null values for each column and percentage of missing values in each column sorted in descending order. 

```python
def missing_values(data):
    Missing_Total = data.isnull().sum().sort_values(ascending = False)
    Percentage = (data.isnull().sum()/data.count()*100).sort_values(ascending = False)
    return pd.concat([Missing_Total, Percentage], axis=1, keys=['Missing_Total', 'Percentage'])
```

```python
missing_values(data)
```

Data contains missing values in salary variable. But before treating these null values we will try to understand the reason behind missing values

### Question 3

#### Identify the outliers in the data is any, without using boxplots. Define a function that calculates the outliers and replaces them with some value for identification. Remember to make these changes in a copy of dataset and not the original data to avoid loss of information. 


```python
cat_var = [ 'gender', 'ssc_b','hsc_s','hsc_b', 'degree_t', 'workex', 'specialisation', 'status']
```

```python
num_var = ['sl_no','ssc_p','hsc_p','degree_p','etest_p', 'mba_p' , 'salary']
```

```python
#Function for calculating the outliers and replacing them with null values in a copy of data to avoid loss of information
def outlier_treat(df, x):
    q75,q25 = np.percentile(df[x], [75,25])
    iqr = q75 - q25
    min1 = q25 - (iqr*1.5)
    max1 = q75 + (iqr*1.5)
    df.loc[df[x] < min1, x] = np.nan
    df.loc[df[x] > max1, x] = np.nan
```

```python
df = data.copy()
for i in num_var:
    outlier_treat(df, i)
```

The outliers are replaced with null values for counting them easily in a copy of data because in this case the outlier are expectional cases and not invalid data. Therefore removing them in original data might lead to loss of information
